```python
import readwin32
```

```python
winfile = './data/test/hinet_evt/U20250618000390_20.evt'
chfile = './data/test/hinet_evt/U20250618000390_20.ch'
```

```python
st, chdict = readwin32.read_win32(winfile, chtbl=chfile)
st1 = st.select(station='n.ncnh*')
st1.detrend()
st1.taper(type='cosine', max_percentage=0.04)
st1.filter('bandpass', freqmin=1, freqmax=2, zerophase=True)
st1.plot()
print()
```
